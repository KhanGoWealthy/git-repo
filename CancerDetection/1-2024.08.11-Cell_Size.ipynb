## HIERARCHICAL CLUSTERING

The scipy.cluster.hierarchy package contains two functions, i.e., linkage() and dendogram() for hierarchical clustering. The linkage() function performs agglomerative clustering and the dendogram() function displays the clusters. Various linkage methods are possible. Here we'll use the ward linkage method that merges clusters so that variance of the clusters is minimized. Other linkage options are:

average
single
comp
lete
The linkage() function returns a linkage matrix with information about clusters. This matrix can be viewed using the dendogram() function. The code below performs clustering using the euclidean metric and displays the clus

## Model tuning will be conducted later onters.

```python
df_dropped = df.drop(df.columns[0], axis=1)
df_dropped.head()
```

```python


# Performs agglomerative clustering using `ward` linkage and `euclidean` metric
hc = linkage(df_dropped, method='ward', metric='euclidean')

# Sets the figure size
fig = plt.figure(figsize=(15, 15))

# Displays the dendogram
# The lambda function sets the labels of each leaf
dn = dendrogram(
    hc,
    leaf_label_func=lambda id: y.values[id],
    leaf_font_size=10)
```



Specific Observations:
Distinct Clusters:

What we see is 80 leafs as we reduced the complexity.  There is some cross over but we can see that Cluster 2 is relatively difference vs the rest of the others. 

There are several distinct clusters that can be observed in the dendrogram. If we were to cut the dendrogram at a height of 60, we might get a number of distinct clusters, each containing data points that are similar to each other.
Hierarchical Merges:

Some clusters only merge at a higher level, indicating that they are quite dissimilar from other clusters. For example, clusters merging at heights above 40 are less similar to other clusters.

Dissimilarity Measure:

The height of the branches gives an idea of the dissimilarity between clusters. Branches that merge at higher heights indicate greater dissimilarity between the clusters.

This will help us interpret results of our other methods K-means and DBSCAN

```python
df1 = data1.copy()

df1 = df1.drop(df1.columns[0], axis=1)
df_dropped.head()
```

```python
from sklearn.cluster import KMeans                # For k-Means
from sklearn.model_selection import GridSearchCV  # For grid search
from sklearn.metrics import silhouette_score      # For metrics and scores
from sklearn.preprocessing import StandardScaler  # For standardizing data



# Separates the class "Cluster" variable in y
y = df1.Cluster

# Removes the y column from df
df1 = df1.drop('Cluster', axis=1)



# Standardizes df1
df1 = pd.DataFrame(
    StandardScaler().fit_transform(df1),
    columns=df1.columns)

# Displays the first 5 rows of df
df1.head()
```

```python
# Sets up the kMeans object
km = KMeans(
    n_clusters=3,
    random_state=123,
    init='k-means++',
    n_init=10)

# Fits the model to the data
km.fit(df1)

# Displays the parameters of the fitted model
km.get_params()
```

```python
# Creates a scatter plot
sns.scatterplot(
    x='Autocorr_Lag_2', 
    y='StdDev',
    data=df1, 
    hue=y,
    style=km.labels_,
    palette=["blue", "orange", "green"])

# Adds cluster centers to the same plot
plt.scatter(
    km.cluster_centers_[:,0],
    km.cluster_centers_[:,1],
    marker='x',
    s=200,
    c='red')
```

## k-MEANS: OPTIMIZE VIA SILHOUETTE SCORES

```python
# Sets up the custom scorer
def s2(estimator,X):
    return silhouette_score(X, estimator.predict(X))

# List of values for the parameter `n_clusters`
param = range(2,10)

# KMeans object
km = KMeans(random_state=0, init='k-means++')

# Sets up GridSearchCV object and stores in grid variable
grid = GridSearchCV(
    km,
    {'n_clusters': param},
    scoring=s2,
    cv=2)

# Fits the grid object to data
grid.fit(df1)

# Accesses the optimum model
best_km = grid.best_estimator_

# Displays the optimum model
best_km.get_params()
```

```python
# Plot mean_test_scores vs. n_clusters
plt.plot(
    param,
    grid.cv_results_['mean_test_score'])

# Draw a vertical line, where the best model is
plt.axvline(
    x=best_km.n_clusters, 
    color='red',
    ls='--')

# Adds labels to the plot
plt.xlabel('Total Centers')
plt.ylabel('Silhouette Score')
```

```python
# Creates a scatter plot
sns.scatterplot(
    x='Autocorr_Lag_2', 
    y='StdDev',
    data=df1, 
    hue=y,
    style=best_km.labels_,
    palette=["blue", "orange", "green","red", "purple"])

# Adds cluster centers to the same plot
plt.scatter(
    best_km.cluster_centers_[:, 0],
    best_km.cluster_centers_[:, 1],
    marker='x',
    s=200,
    c='red')
```

```python
from sklearn.cluster import DBSCAN                # For DBSCAN
import numpy as np                                # For various computations
from sklearn.preprocessing import StandardScaler  # For standardizing data    
from sklearn.neighbors import NearestNeighbors    # For nearest neighbors
```

For DBSCAN clustering, we need two parameters:

min_samples: Minimum number of neighboring points for clustering.
eps: Epsilon neighborhood rad
ius
To find the optimal value of eps havee'll constedruct a plot of distances of min_samples (variable k in code below) nearest neighbors and look for a pronounced bend, also called a "knee." The code below uses the NearestNeighbors object to find the distances and plots them.

```python
k = 7

# Fits a NearestNeighbors object to df1
nn = NearestNeighbors(n_neighbors=k) \
    .fit(df1)

# Finds the distance of k closest neighbors
dist, ind = nn.kneighbors(df1)

# Sorts the distances
dist = np.sort(dist, axis=0)[:,1]

# Plots the distances
plt.plot(dist)

# Draws a horizontal line at the bend
plt.axhline(y=11, color='red', ls='--')
```

```python
# Fits a DBSCAN object to df
# Fits a DBSCAN object to df
db=DBSCAN(eps=11, min_samples=k) \
    .fit(df1)

# Creates a scatter plot to visualize the clusters
sns.scatterplot(
    x='Autocorr_Lag_2', 
    y='StdDev',
    data=df1, 
    hue=y,
    style=db.labels_,
    palette=["blue", "orange", "green","red", "purple"])
```
