## Provincial Shapefile Attributes

### This code reads, analyzes, and plots a shapefile.

- Shapefiles are vector-based regional boundaries for a region. They require GeoPandas library, which is similar to Pandas.
- *NOTE: GeoPandas and Rasterio are not compatibile in the most recent version of Python.*
   
***Recommended Python version: 3.9***

Zimbabwe Boundaries - provincial (adm1). (Humdata, 2023)

Source: https://data.humdata.org/dataset/cod-ab-zwe

```python
import geopandas as gpd

# Load the provincial shapefile
provincial_shapefile = gpd.read_file("zwe_admbnda_adm1_zimstat_ocha_20180911.shp")

# Print the first five rows of the attribute table
provincial_shapefile.head()
```

Explore attributes and metadata.

```python
# Constructing the metadata summary
metadata_summary = {
    "Total Features": len(provincial_shapefile),
    "Geometry Types": provincial_shapefile.geom_type.unique(),
    "Coordinate Reference System (CRS)": provincial_shapefile.crs,
    "Extent of Geometries": provincial_shapefile.total_bounds,
    "Attribute Columns": provincial_shapefile.columns.tolist(),
    "Sample Rows": provincial_shapefile.head(20),
    "Data Types of Columns": provincial_shapefile.dtypes,
    "Missing Data by Column": provincial_shapefile.isnull().sum()
}

# To display the metadata summary, iterate through the dictionary
for key, value in metadata_summary.items():
    print(f"{key}:\n{value}\n")
```

```python
import matplotlib.pyplot as plt

# Plot the provinces
provincial_shapefile.plot()
plt.title("Provincial Boundaries")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
