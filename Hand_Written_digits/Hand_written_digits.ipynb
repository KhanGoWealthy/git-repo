```python
import pandas as pd                                                           # For dataframes
import numpy as np                                                            # For various functions
import matplotlib.pyplot as plt                                               # For plotting functions
import seaborn as sns                                                         # For additional plotting functions
from sklearn.model_selection import train_test_split                         # For train/test splits
```

```python
#https://archive.ics.uci.edu/dataset/80/optical+recognition+of+handwritten+digits
```

```python
df = pd.read_csv("optdigits.csv")

df.head()

df.tail()
```

```python
#Sequetially rename all attribute columns (i.e. pixel data) as P0, P2, etc - Renames the class variable as y


df.columns = ["P" + str(i) for i in range(0, len(df.columns) - 1)] +["y"]
```

```python
df.head()

```

```python


df = df.loc[df.y.isin([1,3,6])]

df.head()
```

```python
X_trn, X_tst, y_trn, y_tst = train_test_split(df.filter(regex='\d'), df.y, test_size=0.3, random_state=1)

trn = X_trn
trn["y"] = y_trn

tst = X_tst
tst["y"] = y_tst
```

```python

fig, ax = plt.subplots(nrows=1, ncols=20, figsize=(15, 3.5), subplot_kw=dict(xticks=[], yticks=[]))

for i in np.arange(20):
    ax[i].imshow(X_trn.to_numpy()[i, 0:-1].reshape(8,8), cmap=plt.cm.gray)
plt.show()
```
